2. Dữ liệu môi trường của một nhà xưởng sản xuất được đo và trình bày trong tập tin đính
kèm tb_tracking.xlsx. Thực hiện các thao tác tiền xử lý sau:

a) Thêm vào tiêu đề các cột theo thứ tự: id (định danh của dòng dữ liệu), device_id (định
danh của thiết bị đo), co_level (nồng độ khí CO), humidity (độ ẩm), temperature
(nhiệt độ), time (thời điểm đo đạc).

```python
import pandas as pd

print("Nguyễn Thành Luân - 6351071043")

df = pd.read_excel("tb_tracking.xlsx", header=None)

df.columns = [
    "id",
    "device_id",
    "co_level",
    "humidity",
    "temperature",
    "time"
]
df.head()
```

b) Do điều chỉnh cảm biến nên dữ liệu đo được bắt đầu từ ngày 03/01/2018 mới chính
xác. Những dữ liệu còn lại không có giá trị, yêu cầu sinh viên loại bỏ.

```python
print("Nguyễn Thành Luân - 6351071043")

df['time'] = pd.to_datetime(df['time'], errors='coerce')

df = df[df['time'] >= "2018-01-03"]

df.head()
```

c) Từ cột time sinh viên bóc tách dữ liệu thành thời điểm đo đạc trong ngày tính theo
phút, đặt tên là minutes.

```python
print("Nguyễn Thành Luân - 6351071043")

df['minutes'] = df['time'].dt.hour * 60 + df['time'].dt.minute

df[['time', 'minutes']].head()
```

d) Vẽ đồ thị phân tán (scatter-plots) lần lượt biểu diễn dữ liệu nồng độ khí CO, độ ẩm
và nhiệt độ theo cột minutes vừa tạo ở yêu cầu trên.

```python
import matplotlib.pyplot as plt

print("Nguyễn Thành Luân - 6351071043")

plt.figure(figsize=(12,4))
plt.scatter(df['minutes'], df['co_level'])
plt.xlabel("Minutes")
plt.ylabel("CO Level")
plt.title("CO Level vs Minutes")
plt.show()

plt.figure(figsize=(12,4))
plt.scatter(df['minutes'], df['humidity'])
plt.xlabel("Minutes")
plt.ylabel("Humidity")
plt.title("Humidity vs Minutes")
plt.show()

plt.figure(figsize=(12,4))
plt.scatter(df['minutes'], df['temperature'])
plt.xlabel("Minutes")
plt.ylabel("Temperature")
plt.title("Temperature vs Minutes")
plt.show()
```
