```python
print("Nguyễn Thành Luân - 6351071043")

subscribe = 'subscribe'
not_subscribe = 'not subscribe'

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

group1 = df[df['marital'] == 'single']
group2 = df[df['marital'] == 'married']

ax = sns.histplot(
    group1[group1['y'] == 'yes'].age.dropna(),
    bins=18,
    label=subscribe,
    ax=axes[0],
    kde=False,
    color='g'
)
sns.histplot(
    group1[group1['y'] == 'no'].age.dropna(),
    bins=40,
    label=not_subscribe,
    ax=axes[0],
    kde=False,
    color='y'
)
ax.legend()
ax.set_title('Single')
ax.set_xlabel('age')
ax.set_ylabel('count')

ax = sns.histplot(
    group2[group2['y'] == 'yes'].age.dropna(),
    bins=18,
    label=subscribe,
    ax=axes[1],
    kde=False,
    color='g'
)
sns.histplot(
    group2[group2['y'] == 'no'].age.dropna(),
    bins=40,
    label=not_subscribe,
    ax=axes[1],
    kde=False,
    color='y'
)
ax.legend()
ax.set_title('Married')
ax.set_xlabel('age')
ax.set_ylabel('count')
```

```python
print("Nguyễn Thành Luân - 6351071043")

if 'housing_num' not in df.columns:
    df['housing_num'] = df['housing'].map({'yes': 1, 'no': 0})

if 'loan_num' not in df.columns:
    df['loan_num'] = df['loan'].map({'yes': 1, 'no': 0})

df['debt_indicators'] = df['housing_num'] + df['loan_num']

df.loc[df['debt_indicators'] > 0, 'no_debt'] = 0
df.loc[df['debt_indicators'] == 0, 'no_debt'] = 1

df['no_debt'] = df['no_debt'].astype(int)

print("Thống kê cột 'no_debt' (1 = không nợ, 0 = có nợ):")
print(df['no_debt'].value_counts())

print("\nThống kê cột 'debt_indicators' (tổng số khoản vay):")
print(df['debt_indicators'].value_counts())
```

Câu 6:

```python
print("Nguyễn Thành Luân - 6351071043")

# Thay thế missing bằng 'unknown'
df['job'] = df['job'].fillna('unknown')
df['education'] = df['education'].fillna('unknown')

df[['job', 'education']].describe()
```

```python
print("Nguyễn Thành Luân - 6351071043")

month_mapping = {
    'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6,
    'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12
}

df['month'] = df['month'].map(month_mapping)
df['month'] = df['month'].astype(int)

df['month'].value_counts()
```

```python
print("Nguyễn Thành Luân - 6351071043")

df = df.drop(['contact'], axis=1)
df = df.drop(['poutcome'], axis=1)
df = df.drop(['duration'], axis=1)
df.head()
```

```python
print("Nguyễn Thành Luân - 6351071043")

mean = df["age"].mean()
std  = df["age"].std()
is_null = df["age"].isnull().sum()

rand_age = np.random.randint(mean - std, mean + std, size=is_null)

age_slice = df["age"].copy()
age_slice[df["age"].isnull()] = rand_age

df["age"] = age_slice.astype(int)

df["age"].isnull().sum()
```

```python
print("Nguyễn Thành Luân - 6351071043")

df['balance'] = df['balance'].fillna(df['balance'].median())

df['balance'].describe()
```

```python
print("Nguyễn Thành Luân - 6351071043")

binary_map = {"yes": 1, "no": 0, "unknown": -1}

df["default"] = df["default"].map(binary_map)
df["housing"] = df["housing"].map(binary_map)
df["loan"]    = df["loan"].map(binary_map)

df[['default','housing','loan']].head()
```
